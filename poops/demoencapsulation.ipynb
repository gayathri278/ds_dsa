```python
class Customer:
    def __init__(self,id,name, address):
        self.id=id
        self.name=name
        self.address=address
    def cust_info(self):
        print(f"id:{self.id} name:{self.name} address:{self.address}")
```

```python
#class Cart with fields: name, code, price, qty, total. Create a system 
class Cart():
    def __init__(self, name, code, price, qty):
        self.name=name
        self. code=code
        self.price=price
        self.qty=qty
    def purchased(self):
        total=self.price*self.qty
        print(f"name:{self.name} code:{self.code} price:{self.price} qty:{self.qty} total:{total}")
c=Cart("Biscuits", 101, 50, 2)
c=Cart("Pencil", 102, 10, 5)
c.purchased()
```
