Tree is a non linear data structure where data will be store in hierarchial order.
Types- binary tree, Binary Search Tree, B+ Tree

```python
class Tree:
    def __init__(self, data):
        self.data=data
        self.left=None
        self.right=None
   
root=Tree(5)
root.right=Tree(8)
root.left=Tree(4)
root.left.left=Tree(3)
root.right.right=Tree(100)
```

```python
print(root.left.left.data)
```

inorder: LDR
preorder: DLR
postorder: LRD

```python
def inorder(root):
       if root:
           inorder(root.left)
           print(root.data)
           inorder(root.right)
inorder(root)
```

```python
def preorder(root):
       if root:
           print(root.data)
           preorder(root.left)
           preorder(root.right)
preorder(root)
```

```python
def postorder(root):
       if root:
           postorder(root.left)
           postorder(root.right)
           print(root.data)
postorder(root)
```
