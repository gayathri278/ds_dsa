```python
#std id, name, result
class Student:
    def __init__(self, id, name, result):
        self.id=id
        self.name=name
        self.result=result
        self.next=None
    def std_info(self):
        print(f"ID : {self.id} Name: {self.name} Result: {self.result}")
```

```python
n=int(input("enter number of stuents: "))
student=[]
```

```python

for i in range (n):
    id=input("Enter ID: ")
    name= input("Enter name: ")
    result=input("Enter result: ")
    s1=Student(id, name, result)
    student.append(s1)
```

```python
for s1 in student:
    s1.std_info()
```

```python
cur=s1.id
s1.id.next=s1.name
s1.name.next=s1.result
s1.result.next=None
while cur:
    print(cur.data)
    cur=cur.next
```
